# THE PROBLEM & GOAL

**Goal:**
- These are the following headers which can be called upon by the receptionist to gather information about a guest:
  - booking_ID: customer code given upon creation of reservation
  - adult_count: How many adults are on the reservaiton
  - children_count: How many children are on the reservation
  - weekend_night_count: The number of weekend nights the guest(s) is staying at the hotel
  - week_nights_count: The number of week nights the guest(s) are staying at the hotel
  - meal_plan_choice: Which meal plan the guest has requested while they stay at the hotel
  - parking_spaces_reserved: Whether or not the guest(s) require parking 
    - [0 = no | 1 = yes]
  - type_of_room_reserved: Which type of room the guest(s) have reserved for their reservation
  - days_reserved_before_arrival: Amount of time between the booking of the reservation and when the reservation will occur
  - year_arrival: The year the guest(s) will arrive for their reservation
  - month_arrival: The month the guest(s) will arrive for their reservation
  - date_arrival: The date the guest(s) will arrive for their reservation
  - date_metric: Numerical value of year, month, and date of reservation
  - market_segment: Description of where reservations originate from
  - previous_guest: Indication of whether or not the guest(s) has made a reservation with the hotel before
      - [0 = no | 1 = yes]
  - previous_cancellations: How many reservations the guest(s) have canceled prior to their current reservation
  - previous_bookings_not_canceled: How many reservations the guest(s) have not canceled prior to their current reservation
  - price_per_room_average: The average amount that the room will cost for the duration of the guest(s) stay
  - special_requests: The total amount of special requests that the guest has made
  - status_of_booking: Shows if the booking has been canceled or not

# CLEANING DATA: FEATURE ENGINEERING

```python
# Display number of unique values by column
dataframe.nunique()
```

```python
# Modify data that has "Not_Canceled" for "status_of_booking" to 0, and "Canceled" to 1
dataframe["status_of_booking"].replace("Not_Canceled", 0, inplace=True)
dataframe["status_of_booking"].replace("Canceled", 1, inplace=True)

# Modify data that has "Not selected" for "meal_plan_choice" to 0
dataframe["meal_plan_choice"].replace("Not Selected", 0, inplace=True)

# Modify data that has "Meal Plan 1" for "meal_plan_choice" to 1
dataframe["meal_plan_choice"].replace("Meal Plan 1", 1, inplace=True)

# Modify data that has "Meal Plan 1" for "meal_plan_choice" to 2
dataframe["meal_plan_choice"].replace("Meal Plan 2", 2, inplace=True)

# Modify data that has "Meal Plan 1" for "meal_plan_choice" to 3
dataframe["meal_plan_choice"].replace("Meal Plan 3", 3, inplace=True)

# Modify data that has "Room_Type 1" for "type_of_room_reserved" to 1
dataframe["type_of_room_reserved"].replace("Room_Type 1", 1, inplace=True)

# Modify data that has "Room_Type 1" for "type_of_room_reserved" to 2
dataframe["type_of_room_reserved"].replace("Room_Type 2", 2, inplace=True)

# Modify data that has "Room_Type 1" for "type_of_room_reserved" to 3
dataframe["type_of_room_reserved"].replace("Room_Type 3", 3, inplace=True)

# Modify data that has "Room_Type 1" for "type_of_room_reserved" to 4
dataframe["type_of_room_reserved"].replace("Room_Type 4", 4, inplace=True)

# Modify data that has "Room_Type 1" for "type_of_room_reserved" to 5
dataframe["type_of_room_reserved"].replace("Room_Type 5", 5, inplace=True)

# Modify data that has "Room_Type 1" for "type_of_room_reserved" to 6
dataframe["type_of_room_reserved"].replace("Room_Type 6", 6, inplace=True)

# Modify data that has "Room_Type 1" for "type_of_room_reserved" to 7
dataframe["type_of_room_reserved"].replace("Room_Type 7", 7, inplace=True)

# Modify unique txt from "market_segment" to numeric
dataframe["market_segment"] = dataframe["market_segment"].map({"Offline": 0, "Online": 1, "Corporate": 2, "Aviation": 3, "Complementary": 4})
```
